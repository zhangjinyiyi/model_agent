```python
import networkx as nx

# 创建一个有向图对象
G = nx.DiGraph()

# 添加节点
G.add_node("A")
G.add_node("B")
G.add_node("C")

# 添加有向边并命名输入和输出
G.add_edge("A", "B", input_name="input1", output_name="output1")
G.add_edge("B", "C", input_name="input2", output_name="output2")
G.add_edge("C", "A", input_name="input3", output_name="output3")

# 打印边的属性
for u, v, data in G.edges(data=True):
    print(f"Edge from {u} to {v} with input name: {data['input_name']} and output name: {data['output_name']}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建一个有向图对象
G = nx.DiGraph()

# 添加节点
G.add_node("A")
G.add_node("B")
G.add_node("C")

# 添加有向边并命名输入和输出
G.add_edge("A", "B", input_name="input1", output_name="output1")
G.add_edge("B", "C", input_name="input2", output_name="output2")
G.add_edge("C", "A", input_name="input3", output_name="output3")

# 绘制图形
pos = nx.spring_layout(G)  # 使用 spring 布局

# 绘制节点
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000)

# 绘制边
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20)

# 绘制节点标签
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

# 绘制边标签
edge_labels = {(u, v): f"{data['input_name']} -> {data['output_name']}" for u, v, data in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

# 显示图形
plt.title("Directed Graph with Named Inputs and Outputs")
plt.axis('off')  # 关闭坐标轴
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建一个有向图对象
G = nx.DiGraph()

# 添加节点
G.add_node("A")
G.add_node("B")
G.add_node("C")

# 添加有向边
G.add_edge("A", "B")
G.add_edge("B", "C")
G.add_edge("C", "A")

# 绘制图形
pos = nx.spring_layout(G)  # 使用 spring 布局

# 绘制节点
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000, alpha=0.7)

# 绘制边，确保有箭头
nx.draw_networkx_edges(G, pos, arrowstyle='-|>', arrowsize=20, edge_color='black', connectionstyle='arc3,rad=0.1')

# 绘制节点标签
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

# 显示图形
plt.title("Directed Graph with Arrows")
plt.axis('off')  # 关闭坐标轴
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# 创建一个有向图
G = nx.DiGraph()

# 添加节点和边
G.add_edges_from([
    ('Start', 'Step 1'),
    ('Step 1', 'Step 2'),
    ('Step 2', 'Step 3'),
    ('Step 3', 'End'),
    ('Step 1', 'Step 4'),
    ('Step 4', 'End')
])

# 使用层次布局
pos = nx.nx_agraph.graphviz_layout(G, prog='dot')

# 绘制图形
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, arrows=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold', edge_color='gray')
plt.title('Flowchart Layout')
plt.show()
```
